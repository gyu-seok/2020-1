```python
#스택 --> 시간초과 --> 강의보기
num = int(input("숫자입력: "))
numlist = []
cnt = -1
for i in range(num):
    com = input("명령입력: ")
    
    if 'push' in com:
        temp = com.split(' ')
        numlist.append(int(temp[1]))
        cnt+=1
    elif com == 'pop':
        if len(numlist) == 0:
            print(-1)
        else:
            print(numlist[cnt])
            numlist.pop(cnt)
            cnt -= 1
    elif com == 'size':
        print(len(numlist))
    elif com == 'empty':
        if len(numlist) == 0:
            print(1)
        else:
            print(0)
    elif com == 'top':
        if len(numlist) == 0:
            print(-1)
        else:
            print(numlist[cnt]) 
```

```python
#단어뒤집기--> 맞았습니다!

num = int(input(""))

def reverse(string):
    result=""
    for i in range(len(string)-1,-1,-1):
        result += string[i]
    return result


for i in range(num):
    result = ""
    sen = input("")
    word = sen.split()
    for j in range(len(word)):        
        result += reverse(word[j])
        
        if j != len(word)-1:
            result = result + " "    
    print(result)        
        
```

```python
#큐 --> 시간초과-->강의보기

import collections

n = int(input())

number=collections.deque([])
for i in range(n):
    command = input()
    temp = command.split()
    if temp[0]== 'push':
        number.append(temp[1])
        
    elif temp[0] == 'pop':
        if len(number) == 0:
            print(-1)
        else:
            print(number[0])
            number.pop(0)
    
    elif temp[0] == 'size':
        print(len(number))
    elif temp[0] == 'empty':
        if len(number) == 0:
            print(1)
        else:
            print(0)
    elif temp[0] == 'front':
        if len(number) == 0:
            print(-1)
        else:
            print(number[0])
            
    elif temp[0] == 'back':
        if len(number) == 0:
            print(-1)
        else:
            print(number[len(number)-1])
```

```python
#스택문제 재시도 --> 실패
T = int(input(""))
global List
List=[]

def push(data):
    List.append(data)
def pop():
    if len(List) == 0:
        print(-1)
    else:
        print(List[len(List)-1])
        List.pop(len(List)-1)
def size():
    print(len(List))
def empty():
    if len(List) == 0:
        print(1)
    else:
        print(0)
def top():
    if len(List) == 0:
        print(-1)
    else:
        print(List[len(List)-1])    
    

i=0
while(i != T):
    cmd = input("")
    cs = cmd.split()
    if cs[0] == 'push':
        push(cs[1])
    elif cmd == "pop":
        pop()
    elif cmd == "size":
        size()
    elif cmd == "empty":
        empty()
    elif cmd == "top":
        top()
            
    i+=1
```
